```python
funcs=[func1,func2,func3,func4,func5,func6,func7,func8,func9,func10,func10,func11,func12,func13,func14,func15,func16,func17,func18,func19,func20,func21,func22,func23,func24]
I=0
eng=['-','0.6 kev','1.0 kev','1.4 kev','1.8 kev','2.2 kev','2.6 kev','3.0 kev','3.4 kev','3.8 kev','4.0 kev','4.4 kev','4.8 kev','5.2 kev','5.6 kev','6.0 kev','6.4 kev','6.8 kev','7.2 kev','7.8 kev','8.2 kev','8.6 kev','9.0 kev','9.4 kev','9.8 kev']
for f in funcs:
    I+=1
    plt.plot(Tim[4500:5300],f[4500:5300],color="k", marker="o", linestyle="-",markersize=1,label=f'Point source {eng[I]} kev)')
    plt.grid(True)
    plt.legend(loc='best',fontsize=4,frameon=True)
    plt.xlabel('Distance along ground (in meters) ')
    plt.ylabel('Normalized intensity')
    plt.title('Simulated diffraction pattern')
    plt.savefig(f'synth_plot_{I}.jpg',dpi=800)
    I=I
    plt.show()
```

```python

def circ (r,x):
    y=np.sqrt(r**2-(x-r)**2)
    return y
size_d=0.1 # size in diameter
size=size_d/2
theta=(size/(1000*3600))*np.pi/180
l=3.384e8
r=(l*mt.tan(theta))
print('Disc radius:',r)
#r=74
x=np.arange(0,(2*r)+1,0.02)
y=circ(r,x)
disc_intensity=((y)/(r))
#print(x)
non_l=np.count_nonzero(~np.isnan(disc_intensity))
print(len(disc_intensity),non_l)
disc_intensity=disc_intensity[:non_l]
xrange=x[:non_l]
print(disc_intensity,xrange)
plt.plot(disc_intensity)
```

```python
convolved_band = convolve1d(cf, disc_intensity)
plt.plot(Tim[4800:5200],convolved_band[4800:5200],color="blue", marker="o", linestyle="-",markersize=1,label=f'Convolved {size_d} milliarcsecond object ')
plt.grid(True)
plt.legend(loc='best',fontsize=4,frameon=True)
plt.xlabel('Distance along ground (in meters) ')
plt.ylabel('Intensity')
plt.title('Simulated diffraction pattern')
plt.savefig(f'conve_{size_d}_0.6-10.jpg',dpi=800)
```
