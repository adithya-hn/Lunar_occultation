```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import fresnel

def Simulate(l,lamda):
    x=np.linspace(5,-5,10000)
    f_scale=np.sqrt(lamda*l)
    u=x/f_scale
    S,C=fresnel(u)
    I=0.5*(0.5+C)**2+0.5*(0.5+S)**2
    #print('Done')
    return (I,x)

```

```python
from astropy import constants as const
from astropy import units as u
#print(const.kpc)
dist=4.9*const.kpc
s_rad=15000000*1000* u.meter  #500km HMBX radius
#print(dist)

angl=np.degrees((s_rad/dist)*u.rad)*3600#*1000

print(angl.value*1000)
```

```python

def circ (r,x):
    y=np.sqrt(r**2-(x-r)**2)
    return y
size_d=0.02 # size in diameter
size=size_d #/2
theta=(size/(1000*3600))*np.pi/180
l=3.384e8
r=(l*theta)#mt.tan(theta))
print('Disc radius:',r)
#r=74
x=np.arange(0,(2*r)+1,0.001)
y=circ(r,x)
disc_intensity=((y)/(r))
#print(x)
non_l=np.count_nonzero(~np.isnan(disc_intensity))
print(len(disc_intensity),non_l)
disc_intensity=disc_intensity[:non_l]
xrange=x[:non_l]
print(disc_intensity,xrange)
plt.plot(disc_intensity)
```
